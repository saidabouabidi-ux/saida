## Exercice 10 (2 pts)

Écrire une fonction qui reçoit une chaîne `mot` et retourne la liste de ses caractères (un caractère par élément) en utilisant une boucle `for`.


```python
def chaine(mot):
    resultat=[]
    for i in range (len(mot)):
        resultat = resultat + [mot[i]]
    return resultat
mot = input("Entrer un mot : ")
resultat=chaine(mot)
print(f"la liste de caractere du mot est : {resultat}")
        
    
    
```
